```python
import time
import pandas as pd
from geopy import Nominatim
geolocator = Nominatim()
import json
from urllib.request import urlopen
```

## Import Dataset

```python
dir = "/home/collier/Dropbox/Skills/Python/Projects/Real_Estate/htx_real_estate_analysis/"
file = pd.read_csv(dir+"all_details_20170922.csv",encoding = "ISO-8859-1")
```

```python
file = file[pd.notnull(file['Zip Code'])]
```

```python
file['zip_str'] = file['Zip Code'].astype(int)
file['zip_str'] = file['zip_str'].astype(str)
```

```python
file['address_full'] = file['Address']+" "+file['City']+" "+file['zip_str']
file['lat'] = ""
file['lon'] = ""
```

```python
file = file[["address_full","lat","lon"]]
```

```python
dir = "/home/collier/Dropbox/Skills/Python/Projects/Real_Estate/walk_score_query_py/"
file_already_scraped = pd.read_csv(dir+"home_addresses_latlon_walkscores_08252017.csv",encoding = "ISO-8859-1")
```

```python
file = file[~file['address_full'].isin(file_already_scraped['address_full'])]
```

```python
file.shape
```

## Collection of Address Locations (Lat&Lon)

```python
for idx,row in file.iterrows():
    try:
        location = geolocator.geocode(row['address_full'])
        file.set_value(idx,'lat',location.latitude)
        file.set_value(idx,'lon',location.longitude)
    except:
        continue
    if idx % 200 == 0:
        print(idx)
    time.sleep(1)
```

## Collection of Walk,Transit,Bike Scores

```python
dir = "C:/Users/David/Dropbox/Skills/Python/Projects/Real_Estate/walk_score_query_py/"
file = pd.read_csv(dir+"file_to_scrape.csv")
```

```python
file['lat'] = file['lat'].astype(str)
file['lon'] = file['lon'].astype(str)
file['bike_score'] = ""
file['transit_score'] = ""
file['walk_score'] = ""
file = file[file['lat']!=""]
```

```python
for idx, row in file.iterrows():
    addr = row['address_full']
    addr = addr.replace(" ","%20")
    lat = row['lat']
    lon = row['lon']
    addr_base = "http://api.walkscore.com/score?format=json&address="
    addr_base = addr_base + addr + "&lat=" + lat + "&lon=" + lon + "&transit=1" + "&bike=1" + "&wsapikey=33a9c4ba9c6f64219f47475fa9e7fd65"
    try:
        jsonurl = urlopen(addr_base)
        text = json.loads(jsonurl.read())
        scores_df = pd.DataFrame.from_dict(text)
        bike_score = scores_df.loc["score","bike"]
        transit_score = scores_df.loc["score","transit"]
        walk_score = scores_df.loc["score","walkscore"]
        file.set_value(idx,"bike_score",bike_score)
        file.set_value(idx,"transit_score",transit_score)
        file.set_value(idx,"walk_score",walk_score)
        del scores_df
    except:
        continue
    if idx % 200 == 0:
        print(idx)
    time.sleep(1)
```
